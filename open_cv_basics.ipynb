```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = Image.open("test.png")
```

```python
img_mat = np.asarray(img)
```

```python
plt.imshow(img_mat)
```

```python
img_1 = cv2.imread("test.png")    //BGR order
```

```python
plt.imshow(img_1)
```

Reason because Blue and Red color are flipped

```python
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_1)
```

```python
resize_img = cv2.resize(img_1,(4000,5000))      #dsize is (width, height)
plt.imshow(resize_img)
```

```python
img_flip = cv2.flip(img_1, 0)   #180degree
```

```python
plt.imshow(img_flip)
```

```python
img_flip = cv2.flip(img_1, 1)
plt.imshow(img_flip)
```
